#Task 7

```python
for i in range(10):
  #print(sum(even[i]))
  mean = sum(even[i]) / len(even[i])
  print('Mean for row ', i, ':', mean)


for j in range(5):
    col_sum = 0
    for i in range(10):
      col_sum += even[i][j]
      col_mean = col_sum / 10
      
    print('Mean for column ', j, ':', col_mean)
```
